## Hopkins Test 
#### Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express no clusterability, a score tending to 1 express an uniform distrubution so clustering can't be useful and  0 express  not an uniform distrubution hence clusterering can be useful.

```python
from pyclustertend import hopkins
```

```python
hopkins(X, X.shape[0])#0 ile 0.5 arasında olması önemli olan 0 a olan yakınlığı data kümelemeye o kadar yakın demek 
#0.5 ile 1 arası uniform dağılıma yakın 
```

### Elbow Method 
### We use this method to find optimal k value by looking up to break point of graph. 

```python
ssd = []
                                       #k sayısını seçmeye geldik elbow 2 ile 10 arrası 
K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)#inertia her kümenin merkeze olan uzaklığını ölçüp karesini alıyor
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") #keskin düşüşün durduğu ilk nokta 
```

```python
ssd # sum of squared distance #en keskin düşüşün durduğu ilk nokta 3 gibi duruyor
```

```python
pd.Series(ssd).diff() #farkları alan bi fonksiyon alttan üstü çıkarıyor
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

# Silhouette analysis

```python
from sklearn.metrics import silhouette_score   #karar verdiğimiz kümeleme başarılımı değilmi diğer kümelere olan uzaklığı ölçüyor 
                                               #modelin kalitesini ölçen bir scorelama 
```

```python
silhouette_score(X, K_means_model.labels_) #ortalama değer
```

```python
range_n_clusters = range(2,9)                       # en iyi scor e2 lakin 2 class için çok iyi puanlama vermiş 
for num_clusters in range_n_clusters:               #teori bize 3 demişti onun siluate scoreda çok yüksek 
    # intialise kmeans                              #2 nin inercia puanı çok yüksek 
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
    #4 e yakın bir değerde anlaşıyoruz 
```

## Hierarchical Clustering

### Dendogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
hc_ward = linkage(y = X, method = "ward")
hc_complete = linkage(X, "complete")
hc_average = linkage(X, "average")
hc_single = linkage(X, "single")
```

```python
plt.figure(figsize = (20,8))
plt.title("Dendrogram")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10);
```

```python
plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size = 10);
```

```python
plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size = 10, truncate_mode='lastp', p=10);
```
